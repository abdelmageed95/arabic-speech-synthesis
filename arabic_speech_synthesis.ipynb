```python
def num_to_arabic_letters(text):
    text = text.split()
    sent = []
    for i in text:
        try:
            num = num2words(int(i), lang='ar')
            sent.append(num)
        except:
            sent.append(i)

    return ' '.join(sent) 
```

```python
def infer_preprocess(input_text):
    modified_line = input_text.replace(',', '')  # Remove commas from the line
    modified_line = modified_line.replace('،', '.')
    modified_line = modified_line.replace('-', ' ')
    modified_line = num_to_arabic_letters(modified_line)
    
    return modified_line
    
```

### Train/ Test Split

```python
import os
import shutil
import random
import pandas as pd



# Set paths

input_csv_path = '/home/ec2-user/SageMaker/speech_synthesis/dataset/dataset_file.csv'
audio_folder_path = '/home/ec2-user/SageMaker/speech_synthesis/dataset/audio'
train_folder = '/home/ec2-user/SageMaker/speech_synthesis/dataset/train'
test_folder = '/home/ec2-user/SageMaker/speech_synthesis/dataset/val'

# Create train and test folders
os.makedirs(train_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# Load CSV file
data = pd.read_csv(input_csv_path)

# Shuffle the data
data = data.sample(frac=1, random_state=42)

# Split data into train and test
split_ratio = 0.95  # 80% for training, 20% for testing
split_index = int(len(data) * split_ratio)
train_data = data[:split_index]
test_data = data[split_index:]

# Copy audio files to train and test folders
for subset_data, subset_folder in [(train_data, train_folder), (test_data, test_folder)]:
    subset_audio_folder = os.path.join(subset_folder, audio_folder_path)
    os.makedirs(subset_audio_folder, exist_ok=True)
    
    subset_csv_path = os.path.join(subset_folder, 'data.csv')
    subset_data.to_csv(subset_csv_path, index=False)
    
    for _, row in subset_data.iterrows():
        audio_filename = row['audio']
        audio_source_path = os.path.join(audio_folder_path, audio_filename)
        audio_dest_path = os.path.join(subset_audio_folder, audio_filename)
        
        # Copy the file only if source and destination paths are different
        if audio_source_path != audio_dest_path:
            shutil.copy(audio_source_path, audio_dest_path)

print("Data split into train and test folders.")


```
